```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import tensorflow as tf
```

```python
dataset = pd.read_csv(r"C:\Users\akshayd6\Desktop\MyModels\ANN - Customer churn\Churn_Modelling.csv")
```

```python
dataset.head()
```

## Encoding Gender, Geography

#### Gender

```python
gender_male = np.where(dataset['Gender'] == 'Male',1,0)
```

```python
dataset['gender_male'] = gender_male
```

#### Geography

```python
geo_dummies = pd.get_dummies(dataset['Geography'],prefix = 'Geo_')
```

```python
dataset = pd.concat([dataset,geo_dummies],axis = 1)
```

```python
dataset.head()
```

```python
dataset.drop(['RowNumber','CustomerId','Surname','Gender','Geography'], axis = 1, inplace = True)
```

```python
dataset.head()
```

```python
X = dataset.drop(['Exited'],axis = 1).values
y = dataset['Exited'].values
```

## Test train split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=101)
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

## ANN

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation,Dropout,Input
```

```python
model = Sequential()
```

```python
np.shape(X_train)
```

```python
model.add(Input(shape = (12,)))
model.add(Dense(units = 10, activation = 'relu'))
model.add(Dense(units = 10, activation = 'relu'))
model.add(Dense(units = 1, activation = 'sigmoid'))
```

```python
model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
model.fit(X_train,y_train, batch_size = 32, epochs = 125)
```

```python
preds = model.predict(X_test)
predictions = pd.DataFrame(preds, columns = ['Predictions'])
```

```python
predictions['Exited'] = np.where(predictions['Predictions']>0.55,1,0)
```

```python
predictions.head()
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
print(classification_report(y_test,predictions['Exited']))
```

```python
print(confusion_matrix(y_test,predictions['Exited']))
```
